```python
from math import sin
import numpy as np
from scipy.integrate import odeint

g = 9.8

def pendulum_equations(w, t, l):
    # dth/dt, dv/dt
    th, v = w
    return np.array([v, - g/l * sin(th)]) 
    
import matplotlib.pyplot as plt
t = np.linspace(0, 10, 1000)
track = odeint(pendulum_equations, (1.0, 0), t, args=(1.0,))
plt.plot(t, track[:, 0], color='g')
plt.plot(t, track[:, 1], color='b')
plt.title("Angle change, init theta=1.0(radian)")
plt.xlabel("t")
plt.ylabel("Seismic angle (radian)")
plt.show()
```

```python
# Damped() pendulum
def pend(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt
b = 0.25
c = 5.0
y0 = [np.pi - 0.1, 0.0]
t = np.linspace(0, 10, 101)
sol = odeint(pend, y0, t, args=(b, c))

import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
from scipy.integrate import odeint 
import numpy as np 

def lorenz(w, t, p, r, b): 
    # wp, r, b
    # dx/dt, dy/dt, dz/dt
    x, y, z = w
    # lorenz 
    return np.array([p*(y-x), x*(r-z)-y, x*y-b*z]) 

t = np.arange(0, 30, 0.01) #  
# odelorenz,  
track1 = odeint(lorenz, (0.0, 1.00, 0.0), t, args=(10.0, 28.0, 3.0)) 
# track2 = odeint(lorenz, (0.0, 1.01, 0.0), t, args=(10.0, 28.0, 3.0)) 

# 
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 

fig = plt.figure()
ax = Axes3D(fig)
ax.plot(track1[:,0], track1[:,1], track1[:,2])
# ax.plot(track2[:,0], track2[:,1], track2[:,2])
plt.show()
```
